```python
"""
accuracy = correct_pred / total_pred

sklearn metrics:- 
model accuracy 
100 95 5

accuracy is not enough when your data is imbalanced.
precision used especially when false positives are dangerous/costly.
recall is used especially when missing case is dangerous/costly.
"""

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# True answers (what actually happens)
y_true = [1, 0 , 1, 1, 0, 1, 0]

# Model predictions (what it guessed)
y_pred = [1, 0, 1, 0, 0, 1, 1]

# evaluating the model
accuracy = accuracy_score(y_true, y_pred)
print ("Accuracy:", accuracy)
precision = precision_score(y_true, y_pred)
print ("Precision:", precision)
recall = recall_score(y_true, y_pred)
print ("Recall:", recall)
f1 = f1_score(y_true, y_pred)
print ("F1:", f1)
```
