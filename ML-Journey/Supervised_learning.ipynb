```python
"""
linear regression
1- finds a pattern in old data 
2- straight line 
3- line 
y = m*x + b

[[value]] 2d list
[[6]]

model sees input as 
X = [[1], [2], [3], [4], [5]]  2d array
so the output is also 2d array

Let training data:- 
[
[1],
[2],
[3]
]
test data:- 
[
[4],
[5] 
]
output data:-
[
[6]
]

In dataframe for model, each row is treated as one sample and each column as one feature
"""

from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]   # input data
y = [[40], [50], [65], [75], [90]]  # output data

model = LinearRegression()

model.fit(X, y)    # function .fit() trains the model

hours = float(input("Enter hours studied: "))

predicted_marks = model.predict([[hours]])     # function .predict() makes a prediction based on the trained model

print(f"Predicted score for studying {hours} hours: {predicted_marks}")
```

```python
from sklearn.linear_model import LogisticRegression

X = [[1], [2], [3], [4], [5]]   # hours studied <--- input data
y = [[0], [0], [1], [1], [1]]   # pass/fail <--- output data

model = LogisticRegression()

model.fit(X, y)

hours = float(input("Enter hours studied: "))

result = model.predict([[hours]]) [0]

if result == 1:
    print(f"Predicted result for studying {hours} hours: you are likely to PASS")
elif (result == 0 and hours < 1):
    print(f"Predicted result for studying {hours} hours: Tui Gaar mara bara abar pass/fail dekhte eseche")
else: 
    print(f"Predicted result for studying {hours} hours: you can FAIL")
```

```python
from sklearn.neighbors import KNeighborsClassifier

X= [
    [190, 7],
    [200, 7.5],
    [250, 8],
    [300, 8.5],
    [330, 9],
    [360, 9.5]
]

# 0 = Apple, 1 = Orange
y = [0, 0, 0, 1, 1, 1]

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X, y)

weight = float(input("Enter weight of the fruit (in grams): "))
size = float(input("Enter size of the fruit (in cm): "))

result = model.predict([[weight, size]])[0]

# That is why we use [0] at the end to get the first element of the array
# result = model.predict([[weight, size]]) --> [0] <-- Its a list 
# result = model.predict([[weight, size]])[0] --> 0 <-- Its an integer

if result == 0:
    print("The fruit is an Apple.")
else:
    print("The fruit is an Orange.")
```
